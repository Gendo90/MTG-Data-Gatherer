```python
import pandas as pd
```

```python
# import dataframe for all cards to csv file
all_cards_df = pd.read_csv("./all_mtg_cards.csv")
all_cards_df
```

## Data Cleaning - Duplicate Removal

```python
# remove duplicates based on name
all_cards_no_dups = all_cards_df.drop_duplicates(subset="name")
all_cards_no_dups.head()
```

## Artist Exploration

```python
# view top 30 most prolific artists in the MTG game
all_cards_no_dups.groupby(['artist']).count().sort_values(by="name", ascending=False)["name"].head(30)
```

## Color Frequency Exploration

```python
print(all_cards_no_dups["color_identity"].count())
all_cards_no_dups["color_identity"].value_counts()
```

#### So it is more common to have all 5 colors in a multicolored card than any 4 colors together by a factor of more than 6 - there are 81 multicolored cards that use all 5 colors, and only 13 cards that use any combination of 4 colors!

```python
# used to convert a string column to a list of the strings
def str_to_list(cell):
    cell = ''.join(c for c in cell if c not in "'[]")
    cell = cell.split(', ')
    return cell

# define an overall card color category - incuding None and Multi
def color_to_category(x):
    try:
        size = len(x)
        if(size == 0):
            return "None"
        elif(size == 1):
            return x[0]
        else:
            return "Multi"
    except:
        return "None"

all_cards_no_dups["overall_color"] = all_cards_no_dups["color_identity"].apply(str)
all_cards_no_dups["overall_color"] = all_cards_no_dups["overall_color"].apply(lambda x: "[]" if x == "nan" else x)
all_cards_no_dups["overall_color"] = all_cards_no_dups["overall_color"].apply(eval)
all_cards_no_dups["overall_color"] = all_cards_no_dups["overall_color"].apply(color_to_category)
all_cards_no_dups.head()
```

```python
# plot the counts of each card color over the whole dataframe
all_cards_no_dups.groupby(['overall_color']).count().plot(kind='bar', y='name')
```

#### Most of the colors, including multicolor, have approximately the same number of cards, but there are far fewer cards with no color than any given color (only about 70% as many)

## Subtype Exploration

```python
all_cards_no_dups.iloc[0]["subtypes"]
```

```python
# show counts of creature types
all_cards_no_dups["subtypes"] = all_cards_no_dups["subtypes"].apply(str)
all_cards_no_dups["subtypes"] = all_cards_no_dups["subtypes"].apply(lambda x: "[]" if x == "nan" else x)
all_cards_no_dups["subtypes"] = all_cards_no_dups["subtypes"].apply(eval)
all_cards_no_dups["subtypes"].value_counts()
```

```python
card_color_subtypes = all_cards_no_dups.loc[:, ["name", "subtypes", "overall_color"]]
card_color_subtypes = card_color_subtypes.explode("subtypes")
card_color_subtypes.head()
```

```python
card_subtype_count_by_color = card_color_subtypes.groupby(["subtypes", "overall_color"]).count()
card_subtype_count_by_color
```

```python
# print top 30 most common black subtypes

black_types = card_subtype_count_by_color.loc[card_subtype_count_by_color.index.get_level_values("overall_color") == "B"]
black_types = black_types.sort_values(by="name", ascending=False)
print(black_types.head(30))
black_types.head(30).plot(kind="bar")
```

#### So the most common subtype for black cards is, surprisingly enough, Zombie! They just edged out humans by 13 cards, which is the second most frequent black subtype by far!

```python
# print top 30 most common multicolor subtypes

multi_types = card_subtype_count_by_color.loc[card_subtype_count_by_color.index.get_level_values("overall_color") == "Multi"]
multi_types = multi_types.sort_values(by="name", ascending=False)
print(multi_types.head(30))
multi_types.head(30).plot(kind="bar")
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
# define word map generator function
def make_word_map(col, df):
    comment_words = ''
    stopwords = set(STOPWORDS)

    # iterate through the csv file
    for val in df[col]:

        # typecaste each val to string
        val = str(val)

        # split the value
        tokens = val.split()

        # Converts each token into lowercase
        for i in range(len(tokens)):
            tokens[i] = tokens[i].lower()

        comment_words += " ".join(tokens)+" "

    wordcloud = WordCloud(width = 800, height = 800,
                    background_color ='white',
                    stopwords = stopwords,
                    min_font_size = 10).generate(comment_words)

    # plot the WordCloud image                      
    plt.figure(figsize = (8, 8), facecolor = None)
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.tight_layout(pad = 0)

    plt.show()
```

### Card Name Text Word Map

```python
make_word_map("name", all_cards_no_dups)
```

### Card Rule Text Word Map

```python
make_word_map("original_text", all_cards_no_dups)
```

### Card Flavor Text Word Map

```python
make_word_map("flavor", all_cards_no_dups)
```

## Rulings

Which cards have caused the most judge calls - or at least, needed to be clarified the most?

```python
most_rulings = all_cards_no_dups.loc[:, ["name", "rulings"]]
most_rulings.head()
```

```python
def get_ruling_count(x):
    try:
        return len(x)
    except:
        return 0

most_rulings["count"] = most_rulings["rulings"].apply(get_ruling_count)
most_rulings = most_rulings.sort_values(by="count", ascending=False)
most_rulings.head(30)
```
